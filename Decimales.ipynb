### Este código sirve para encontrar el período de la expansión decimal de fracciones $\frac{1}{p}$, con $p$ primo. 

```python
from decimal import *
import math
```

```python
a=1                 # El módulo "decimal" sirve para controlar la cantidad de decimales mostrados en los cálculos
b=47

getcontext().prec = 3*b
fraccion = Decimal(a) / Decimal(b)
print(fraccion)
```

```python
def find_period(n, d):                 # Esta función determina el periodo de fracciones                       
    z = x = n * 9                      # propias e irreducibles con denominador primo
    k = 1
    while z % d:
        z = z * 10 + x
        k += 1

    digits = f"{z // d:0{k}}"
    return k, digits

# Test

num, den = a, b
period, digits = find_period(num, den)
print('num:', num, 'den:', den, 'period:', period, 'digits:', digits)
```

```python
from tabulate import tabulate       # Usamos el módulo "tabulate" para hacer más bonita la tabla

lista=[]

for p in primos:
    if p<100:
        longitud, periodo=find_period(1, p)
        lista.append([p, longitud, periodo])
        
    else:
        continue
```
